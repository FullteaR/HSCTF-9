```python
w10 = [w.strip() for w in open("10.txt", "r")]
w4 = [w.strip() for w in open("4.txt", "r")]
w2 = [w.strip() for w in open("2.txt", "r")]
```

```python
def convert(word):
    if len(set(word[0]+word[2:6]+word[7:]))==len(set("TAXYZFUE")):
        return
    m = {}
    for e,c in zip("T1AXYZ6FUE", word):
        m[e] = c
    
    message = ""
    message_origin = ""
    for char in cipher:
        if char in m:
            message+='\033[32m'+m[char]+'\033[0m'
            message_origin+=m[char]
        else:
            message+='\033[31m'+char+'\033[0m'
            message_origin+=char
    if message_origin.split(" ")[12] in w2 and message_origin.split(" ")[22] in w4:
        print(message)
```

```python
for word in tqdm(w10):
    convert(word)
```
